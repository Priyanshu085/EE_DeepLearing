```python
q_table = np.shape((48, 4))

EPSILON= 0.1
ALPHA = 0.1
GAMMA = 0.9
NUM_EPISODES = 500
```

```python
def policy(state, explore=0.0):
  action = int(np.argmax(q_table[state]))
  if np.random.rand() <= explore:
    action = int(np.random.randint(low=0, high=4, size=1))
  return action
```

# Q-Learning Implementation

```python
q_table = np.zeros((48, 4))

for episode in range(NUM_EPISODES):
  done= False
  total_reward= 0
  episode_length= 0

  state = cliffEnv.reset()
  while not done:
    action = policy(state, EPSILON)
    next_state, reward, done, _ = cliffEnv.step(action)[:4]
    next_action = int(policy(next_state))  # Convert next_action to an integer

    q_table[state][action] += ALPHA * (reward + GAMMA * q_table[next_state][next_action] - q_table[state][action])

    state = next_state

    total_reward += reward
    episode_length += 1

```
